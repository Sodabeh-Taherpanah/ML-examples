### Plot with Different Interfaces
Plot the function $ y = sin(x) * cos(x) $ using both MATLAB-style and object-oriented interfaces. Decorate the plots properly and save them as png files.

```python
import numpy as np
import os
item = np.linspace(-2 * np.pi, 2 * np.pi, 400)
def yFunc(x):
   return  np.sin(x) * np.cos(x)

z=yFunc(item)

fig, ax = plt.subplots(figsize=(9,6))


ax.plot(item,z,label='hi',color='g', linestyle='--')
ax.set_xlabel('phase')
ax.set_ylabel('value')
ax.set_title("Our Plot: y = sin(x) * cos(x)")
ax.grid(True)
ax.legend()


# Save the figure as a PNG file
fig.savefig("as.png", dpi=300)

# Show the plot
plt.show()


#print("Current  directory:", os.getcwd())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()

# converts the dataset into a Pandas DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Calculate the mean and standard deviation for each feature
means = df.mean()
#the average measurements of the Iris flower features
print(means)

std_devs = df.std()

# bar plot with error bars
fig, ax = plt.subplots(figsize=(9, 6))

# Plotting the bar chart with error bars. ,yerr=std_devs ,
# add small cap to the ends of error bars to show better

ax.bar(means.index, means, yerr=std_devs, capsize=5,
       color='lightgreen', edgecolor='black',
       error_kw={'elinewidth': 3, 'ecolor': 'red'})  # Error bars

# Add labels and title
ax.set_xlabel('Features*')  # x-axis label
ax.set_ylabel('Values')  # y-axis label
ax.set_title('Mean and Standard Deviation of Iris Dataset Features')  # Title


plt.xticks(rotation=0)  # for better visibility. if you want you can Rotate x-axis : 45 labels
plt.tight_layout()  # avoid overlapping with Adjust layout
plt.show()


```

### Draw Subplots
Plot a 2x2 grid. Each subplot of this grid should draw the histogram of one feature of the iris dataset. Decorate the plot properly.

```python
import sklearn.datasets
iris = sklearn.datasets.load_iris()

```

```python
# converts the dataset into a Pandas DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plotting the bar chart with error bars. ,yerr=std_devs ,
# add small cap to the ends of error bars to show better
colors = ['skyblue', 'salmon', 'lightgreen', 'orange']

# zip is
# A subplot (ax) from the 2×2 grid.
# A feature name (feature: feature_item) from the Iris dataset columns.

for ax, feature_item, color in zip(axes.flat, df.columns, colors):
    ax.hist(df[feature_item], bins=15, color=color, edgecolor='black', alpha=0.7)
    # alpha: Adjusts transparency.

    ax.set_title(f"Histogram:  {feature_item}")
    ax.set_xlabel("Value")
    ax.set_ylabel("Frequency")
    ax.grid(True)


plt.xticks(rotation=0)  # for better visibility. if you want you can Rotate x-axis : 45 labels
plt.tight_layout()  # avoid overlapping with Adjust layout
plt.show()

```

### Draw a 3D Plot
Visualize the data points of the iris dataset in a 3D space using the first three features. Decorate the figure properly.

```python
import sklearn.datasets
iris = sklearn.datasets.load_iris().data
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()

# Convert to DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Extract first three features
x = df.iloc[:, 0]  # Sepal Length
y = df.iloc[:, 1]  # Sepal Width
z = df.iloc[:, 2]  # Petal Length
labels = iris.target  # Target class labels

# Define colors for different classes
colors = ['red', 'green', 'blue']
class_names = iris.target_names  # ['setosa', 'versicolor', 'virginica']

# Create a 3D figure
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with colors based on class labels
for i in range(3):
    ax.scatter(x[labels == i], y[labels == i], z[labels == i],
               label=class_names[i], color=colors[i], alpha=0.7, edgecolors='black')

# Decorate the plot
ax.set_xlabel("Sepal Length (cm)")
ax.set_ylabel("Sepal Width (cm)")
ax.set_zlabel("Petal Length (cm)")
ax.set_title("3D Visualization of the Iris Dataset")
ax.legend()

# Show the plot
plt.show()
```
